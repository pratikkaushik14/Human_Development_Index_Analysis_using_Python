# Introduction

The chosen subject matter of this project revolves around the analysis of Human Development Index (HDI) and its relationship with various societal factors. The HDI is a summary measure of human development that was introduced in 1990 by the United Nations Development Programme (UNDP) [1]. The UNDP evaluates and ranks countries based on three fundamental dimensions of human development: longevity, education and standard of living. Longevity is measured by life expectancy at birth, reflecting the overall health and longevity of a nation's population. The education dimension is measured through two metrics: expected years of schooling for children entering schools and the mean years of schooling for the adult population. The standard of living is measured using the Gross National Income (GNI) per capita. #### SQL Query Q1: 

```python
#Guiding Question 1: How has the Human Development Index (HDI) changed over the past three decades across different countries and continents?


#Computing the HDI change for each decade and grouping by continents

#Use AVG function to compute the AVG HDI for the years 1990,2000,2010 and 2020
avg_hdi_change_query="""
SELECT 
    c.Continents,
    AVG(h.`HDI 1990`) as Avg_HDI_1990,
    AVG(h.`HDI 2000`) as Avg_HDI_2000,
    AVG(h.`HDI 2010`) as Avg_HDI_2010,
    AVG(h.`HDI 2020`) as Avg_HDI_2020,
    ((AVG(h.`HDI 2020`) - AVG(h.`HDI 1990`)) / AVG(h.`HDI 1990`)) * 100 as Avg_HDI_Percentage_Change_1990_to_2020
FROM 
    hdi_table2 h
INNER JOIN 
    country_continents c ON h.Country = c.Country
GROUP BY 
    c.Continents
ORDER BY 
    Avg_HDI_Percentage_Change_1990_to_2020 DESC;
"""
hdi_change_by_continents=pd.read_sql_query(avg_hdi_change_query,engine)
hdi_change_by_continents.head(6)
```

#### Inference: 
In addressing the first guiding question, we employed the AVG function to calculate the average Human Development Index (HDI) values for the years 1990, 2000, 2010, and 2020. Subsequently, we determined the percentage change in HDI from 1990 to 2020, categorizing the average HDI values by continent.

Analyzing the resultant table reveals a discernible upward trend in HDI values across all continents throughout different decades. Notably, Africa and Asia exhibited the most substantial changes in HDI values from 1990 to 2020. Specifically, Africa experienced a remarkable 32.32% increase, while Asia and South America followed closely with 26.88% and 20.54%, respectively. In contrast, Oceania, North America, and Europe demonstrated relatively modest changes, averaging around 17%. Noteworthy is Europe's consistent status with the highest average HDI value across decades, in contrast with Africa's persistently lower averages. #### Part 1: What is the correlation between the Human Development Index (HDI) and Gender Inequality Index (GII) across all countries?

The SQL query below calculates the correlation coefficient between the Gender Inequality Index (GII) and the Human Development Index (HDI) components from the given tables (hdi_table1 and hdi_table5), with tables being joined by country.

```python
# Correlation_coefficient = Covariance(x,y)/STD(x)*STD(y)
# Covariance(x,y) = AVG((x-x.mu)*(y-y.mu))

corr_gii_hdi = '''

-- Calculate the mean
WITH Mean AS(
SELECT 
        gender_inequality_index_value AS x,
        hdi_index_components AS y,
        AVG(gender_inequality_index_value) OVER() AS mean_x,
        AVG(hdi_index_components) OVER() AS mean_y
FROM
        hdi_table1
JOIN
        hdi_table5
ON
        hdi_table1.country = hdi_table5.Country
),

-- Calculate Covariance and Denominator
Correlation AS(
SELECT
        AVG((x - mean_x)*(y- mean_y)) AS Covariance,
        STDDEV(x)*STDDEV(y) AS Denominator
FROM 
        Mean
)

-- Calculate correlation coefficient
SELECT 
        Covariance/Denominator AS Correaltion_Coefficient
FROM 
        Correlation;
'''
```

```python
# Execute the query and fetch results into a DataFrame
gii_statistic_df = pd.read_sql_query(corr_gii_hdi, engine)

#display
gii_statistic_df
```

Explanation:

Mean CTE (WITH Mean AS ...):
It calculates the mean (average) of GII (Gender_Inequality_Index_Value) and HDI (hdi_index_components) for each corresponding country.AVG(Gender_Inequality_Index_Value) OVER() and AVG(hdi_index_components) OVER() compute the overall mean for GII and HDI.

Correlation CTE (Correlation AS ...):
It calculates the Covariance and Denominator needed for the correlation coefficient formula.
Covariance is calculated using the formula AVG((x - mean_x)*(y- mean_y)).
Denominator is calculated as the product of the standard deviations (STDDEV) of GII and HDI.

Main Query (SELECT Covariance/Denominator AS Correaltion_Coefficient FROM Correlation;):
It computes the correlation coefficient using the formula Correlation/Denominator.
The result represents the strength and direction of the linear relationship between GII and HDI.

Note:
The query employs Common Table Expressions (CTEs) for a structured and modular approach.
The calculation of the correlation coefficient provides insights into how GII and HDI are related.
The correlation coefficient ranges from -1 to 1, where -1 indicates a perfect negative correlation, 1 indicates a perfect positive correlation, and 0 indicates no correlation.

#### Part 2:  How gender inequality gap affects the per capita income of females and males

The SQL query below calculates the correlation coefficient between the Gender Inequality Index (GII) and the Per capita income of males and females separately from the given tables (hdi_table4 and hdi_table5), with tables being joined by country.

```python
# Correlation_coefficient = Covariance(x,y)/STD(x)*STD(y)
# Covariance(x,y) = AVG((x-x.mu)*(y-y.mu))

corr_gii_hdi = '''

-- Calculate the mean
WITH Mean AS(
SELECT 
        gender_inequality_index_value AS x,
        `Estimated gross national income per capita Female(2017 PPP $)` AS y_female,
        `Estimated gross national income per capita Male(2017 PPP $)` AS y_male,
        AVG(gender_inequality_index_value) OVER() AS mean_x,
        AVG(`Estimated gross national income per capita Female(2017 PPP $)`) OVER() AS mean_y_feamle,
        AVG(`Estimated gross national income per capita Male(2017 PPP $)`) OVER()  AS mean_y_male
FROM
        hdi_table4
JOIN
        hdi_table5
ON
        hdi_table4.country = hdi_table5.Country
),

-- Calculate Covariance and Denominator
Correlation AS(
SELECT
        AVG((x - mean_x)*(y_female - mean_y_feamle)) AS Covariance_Female,
        AVG((x - mean_x)*(y_male - mean_y_male)) AS Covariance_Male,
        STDDEV(x)*STDDEV(y_female) AS Denominator_Female,
		STDDEV(x)*STDDEV(y_male) AS Denominator_Male
FROM 
        Mean
)

-- Calculate correlation coefficient
SELECT 
        Covariance_Female/Denominator_Female AS Correaltion_Coefficient_Female,
        Covariance_Male/Denominator_Male AS Correaltion_Coefficient_Male
FROM 
        Correlation;
'''
```

```python
# Execute the query and fetch results into a DataFrame
gii_statistic_df = pd.read_sql_query(corr_gii_hdi, engine)
gii_statistic_df
```

Explanation:

Mean CTE (WITH Mean AS ...):
This part of the query calculates the mean values for GII, female income, and male income. It creates a Common Table Expression (CTE) named 'Mean,' extracting necessary columns from 'hdi_table4' and 'hdi_table5' and calculating the means for GII, female income, and male income using the AVG() function over the entire dataset.

Correlation CTE (Correlation AS ...):
The 'Correlation' CTE calculates the Covariance and Denominator for both female and male income. It computes the Covariance using the formula AVG((x - mean_x)*(y - mean_y)), where x represents GII, and y represents either female or male income. The Denominator is then determined as the product of the standard deviations (STDDEV) of GII and female/male income.

Main Query (SELECT Covariance/Denominator AS Correlation_Coefficient_Female...):
The main part of the query calculates the correlation coefficient for both female and male incomes. It divides the Covariance by the Denominator, providing the correlation coefficient for the relationship between GII and estimated gross national income per capita for females and males, respectively.

Note:
This SQL query utilizes CTEs for a structured and modular approach. The correlation coefficient helps quantify the degree and direction of the relationship between GII and estimated gross national income per capita for both genders. A correlation coefficient close to -1 or 1 suggests a strong correlation, while close to 0 indicates a weaker correlation.

## Are carbon dioxide emissions and material footprint a good indicator of life expectancy and HDI Index across different continents?

#### Importance:

By examining carbon dioxide emissions and material footprint, this question assesses how a countrys environmental impact may correlate with the health and longevity of its population, signified by life expectancy. Similarly, it may provid a potential reflection on a country's development level, as indicated by the HDI. This inquiry is crucial for understanding whether countries with lower environmental impacts achieve better human development outcomes.

#### Definitions: 

* Carbon dioxide emissions per capita: refers to the total CO2 emissions produced from fossil fuels in tonnes,  divided by the countrys population.

* Material footprint per capita: accounts for the raw materials extracted to meet a country's consumption needs in tonnes, divided by the countrys population.

### Query 1:

```python
query_q3A = """
SELECT 
    country_continents.Continents,
    AVG(hdi_table7.HDI_index) AS avg_hdi_index,
    AVG(hdi_table1.life_expectancy_at_birth) AS avg_life_expectancy,
    AVG(hdi_table7.co2_per_capita) AS avg_co2_per_capita,
    AVG(hdi_table7.footprint_per_capita) AS avg_footprint_per_capita
FROM 
    hdi_table7
INNER JOIN 
    hdi_table1 
ON 
    hdi_table7.country = hdi_table1.Country
INNER JOIN 
    country_continents 
ON 
    hdi_table7.country = country_continents.Country
GROUP BY 
    country_continents.Continents
"""

co2_footprint_capita = pd.read_sql(query_q3A, engine)
co2_footprint_capita
```

### Query 1 Explanation:
The SQL query in question is designed to aggregate and correlate data across different continents by joining three distinct datasets: environmental indicators, life expectancy, and continental classifications. The query employs INNER JOIN operations to merge hdi_table7, which contains environmental data like CO2 emissions and material footprint per capita, with hdi_table1, holding life expectancy data, based on a common country identifier. It further associates these records with continental data from country_continents table. Using the AVG function, the query calculates the average Human Development Index (HDI), life expectancy, CO2 emissions, and material footprint for each continent. This is achieved by grouping the results by continent, ensuring that the averages reflect regional, rather than global, statistics.
